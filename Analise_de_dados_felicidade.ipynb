# **Preparação de Dados;**

**Motivação: **
- Dados incompletos - Faltam valores (EX:coluna idade está em branco em uma ou mais linhas)

- Dados ruidosos - são erros ou outliers (EX: idade = 180)

- Dados inconsistentes - discrepância em códigos e nomes (EX: na coluna grau de felicidade ao invés de valores de 0.0 á 10.0 apareceu data)


# **Origem dos Prolemas:**

- Problemas de Hardware;
- Problemas de Solftware;
- Problemas por obter dados de fontes diferentes;
- Problemas causados por interferência humana.


# **Solução dos Problemas:**

- Substituição dos valores: pela média dos demais que atendam uma certa condição ou por um valor padrão;

- Remoção de linhas;
- Remoção de colunas;
- Padronização;
- Normalização;
- Codificação;
